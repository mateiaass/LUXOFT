```python
from collections import Counter
import os

def parse_conllup(files_directory):
    date_counter = Counter()
    i = 0

    # Iterate over files in the directory
    for filename in os.listdir(files_directory):
        # i += 1
        # print(i)
        
        # Construct the full file path
        file_path = os.path.join(files_directory, filename)
        
        # Open and read the file
        with open(file_path, 'r', encoding='utf-8') as file:
            conllup_text = file.read()

        # Split the CoNLLUP text into lines
        lines = conllup_text.strip().split('\n')
        text_start = False
        
        # Iterate over each line
        for line in lines:
            line = line.strip()

            if line.startswith('# sent_id'):
                text_start = True
                continue
                
            if line.startswith('#') and text_start == False:
                key_value = line[1:].strip().split('=', 1)
                if len(key_value) == 2:
                    key, value = key_value
                    if key.strip() == 'date':
                        date_counter[value.strip()] += 1    

    return date_counter
```

```python
files_directory = 'ro-annotated/'

date_counter = parse_conllup(files_directory)

# # Print metadata
# print("Metadata:")
# for key, value in metadata.items():
#     print(f"{key}: {value}")

# Print the frequency of files by date
print("Frequency of files by date:")
for date, frequency in date_counter.items():
    print(f"{date}: {frequency}")
```
